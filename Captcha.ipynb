```python
import urllib
import os

def download(url, name, extension, override=False):
    '''Downloads and saves an image if it dosen't already exsist
    '''
    if not os.path.isfile('data/%s.%s' % (name, extension)) or override:
        with open('data/%s.%s' % (name, extension), 'wb') as image_file:
            image_file.write(urllib.urlopen(url).read())
            image_file.close()
            
download('http://pastebin.com/etc/CaptchaSecurityImages.php?width=1600&height=800&characters=2&b=%27.md5(3).%27', 'captcha', 'jpg')
```

```python
import cv2
import matplotlib.pyplot as plt
from skimage.filter import threshold_adaptive
from skimage.io import imsave
import numpy as np

img = cv2.imread('data/captcha.jpg')

original = img.copy()

plt.figure()
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))

# Apply slighy blur to reduce noise
blur = cv2.GaussianBlur(original, (5, 5), 0)

plt.figure()
plt.imshow(cv2.cvtColor(blur, cv2.COLOR_BGR2RGB))

height, width, depth = img.shape

# crop
crop = blur[100:height-100, 300:width-300]

plt.figure()
plt.imshow(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB))

# gray
gray = cv2.cvtColor(crop, cv2.COLOR_BGR2GRAY)
plt.figure()
plt.imshow(gray, cmap='gray')

# thresh
thresh = threshold_adaptive(gray, 250, offset = 3)
thresh = thresh.astype("uint8") * 255


plt.figure()
plt.imshow(thresh, cmap='gray')

plt.show()

cv2.imwrite('data/captcha_clean.jpg', thresh)
```

```python
def solve(path, debug=False, show_original=True):
    
    # Open the image
    img = cv2.imread(path)
    
    if debug or show_original:
        plt.figure()
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        
    # Apply a slight blur to reduce noise
    img_blur = cv2.GaussianBlur(img, (5, 5), 0)
    
    if debug:
        plt.figure()
        plt.imshow(cv2.cvtColor(img_blur, cv2.COLOR_BGR2RGB))
        
    height, width, depth = img.shape

    # crop, text always in the center, assumes width=1600&height=800
    # TODO: bounding box could be tighter
    img_crop = img_blur[90:height-90, 400:width-400] # 2 letter
    #img_crop = img_blur[90:height-90, 300:width-300]  # 4 letter
    
    if debug:
        plt.figure()
        plt.imshow(cv2.cvtColor(img_crop, cv2.COLOR_BGR2RGB))
        
    # Convert to grayscale
    img_gray = cv2.cvtColor(img_crop, cv2.COLOR_BGR2GRAY)
    
    if debug:
        plt.figure()
        plt.imshow(img_gray, cmap='gray')
        
    # Apply adaptive threshold from skimage
    img_thresh = threshold_adaptive(img_gray, 250, offset = 3)
    img_thresh = img_thresh.astype("uint8") * 255 # Convert back to opencv
    
    if debug:
        plt.figure()
        plt.imshow(img_thresh, cmap='gray')

        
    # Save the image, this is easier than converting to PILs format
    cv2.imwrite('data/captcha_clean.jpg', img_thresh)
    
    tools = pyocr.get_available_tools()
    tool = tools[0] # tesseract
        
    # Use tesseract for ocr, scale the image down which gives
    # better detection
    a = Image.open('data/captcha_clean.jpg')
    a = a.resize((40, 31), Image.ANTIALIAS)
    text = tool.image_to_string(a,
                               lang='eng',
                               builder=pyocr.builders.TextBuilder())
    
    if debug:
        plt.figure()
        plt.imshow(np.asarray(a), cmap='gray')
    
    return text # image_to_string returns unicode
```

```python
solve('data/captcha.jpg')
```

```python
def get_dataset(num=5):
    url = 'http://pastebin.com/etc/CaptchaSecurityImages.php?width=1600&height=800&characters=2&b=%27.md5(3).%27'
    for i in range(num):
        download(url, 'captcha-%i' % i, 'jpg', True)
        
dataset_size = 5
get_dataset(dataset_size)
for i in range(dataset_size):
    print '%i) ' % i + solve('data/captcha-%i.jpg' % i)
    
plt.show()
```
